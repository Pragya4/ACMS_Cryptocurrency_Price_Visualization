```python
@app.route('/BTCDayGraph',methods=['GET','POST'])
def BTCDay():
    cur.execute("select symbol,price_usd,last_updated from public.dummy where symbol='BTC';")
    rows=cur.fetchall()
    BTCd_x_axis_data=[]
    BTCd_y_axis_data=[]
    for i in range(len(rows)-48,len(rows)):
        tupl= rows[i]
        var= tupl[2]
        if type(var)==float:
            BTCd_x_axis_data.append(tupl[2])
            BTCd_y_axis_data.append(tupl[1])
            print tupl[2],tupl[1]
    #conn.close()
    chart=pygal.Line(x_label_rotation=45)
    chart.x_labels=BTCd_x_axis_data
    chart.add('BTC DAY',BTCd_y_axis_data)
    chart=chart.render_data_uri()
    return render_template( 'charts.html', chart = chart)
    
        
            
```
